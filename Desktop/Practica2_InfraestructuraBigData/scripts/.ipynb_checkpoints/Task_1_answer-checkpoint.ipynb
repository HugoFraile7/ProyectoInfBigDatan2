```python
import pandas as pd
import matplotlib.pyplot as plt
from utils import download_dataframe_from_minio

df = download_dataframe_from_minio(
    bucket_name='access-zone',
    object_name='trafico/trafico_congestion_por_hora.parquet',
    format='parquet'  # ← ¡aquí el cambio importante!
)

df.head()
```

```python


TOP_N      = 5


if "Unnamed: 0" in df.columns:
    df = df.drop(columns="Unnamed: 0")

vehicle_cols = ["coches", "motos", "camiones", "buses"]

top_hours = (
    df.assign(vehiculo_predominante=lambda x: x[vehicle_cols].idxmax(axis=1))
      .sort_values("nivel_congestion", ascending=False)
      .head(TOP_N)
      [["hora", "nivel_congestion", "total_vehiculos", "vehiculo_predominante"]]
)

print(f"\nTop {TOP_N} horas con mayor congestión")
print(top_hours.to_string(index=False))

df_sorted = (
df.assign(hora_int=df["hora"].apply(lambda t: t.hour)).sort_values("hora_int")
)

plt.figure()
df_sorted.plot(kind="bar", x="hora", y="total_vehiculos", legend=False,
               title="Total de vehículos por hora")
plt.xlabel("Hora del día")
plt.ylabel("Total de vehículos")
plt.tight_layout()
plt.show()

plt.figure()
df_sorted.plot(kind="bar", x="hora", y="nivel_congestion", legend=False,
               title="Nivel de congestión por hora")
plt.xlabel("Hora del día")
plt.ylabel("Índice de congestión")
plt.tight_layout()
plt.show()

```
